### 4.-   Diagrama de pastel para comparar la cantidad de cervezas nacionales e importadas (0: Nacionales, 1: Importadas).

- 4.1.- Remplazamos los valores numericos con texto para que tenga mas sentido el pie.

```python
df.Importada = df.Importada.replace(1, 'Importadas')
df.Importada = df.Importada.replace(0, 'Nacionales')
```

- 4.2.- Creamos el pie!

```python
df['Importada'].value_counts().plot(kind = 'pie', title = 'Comparación de Cervezas');
```

### 5.- Diagrama de barras para mostrar los tipo de cerveza.

```python
df['Tipo'].value_counts().plot(kind='barh');
```

### 6.- Histograma que muestra la distribución del "costo por 100 mililitros".

```python
# Importa pyplot
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
h=ax.hist(df['Costo100ml'], bins=10, color='skyblue', edgecolor='white', cumulative=False)
ax.set_title('Distribucion del costo por 100 mililitros')
plt.show();
```

### 7.- Diagrama de dispersión para mostrar la relación entre calorías y alcohol.

```python
fig, ax = plt.subplots() #genera el espacio
ax.scatter(df['Calorias'], df['Alcohol']) #grafico diagrama de dispercion 
ax.set_title('Relacion entre Calorias y alcohol') #nombre de las etiquetas del grafico
ax.set_xlabel('Calorias') #eje X
ax.set_ylabel('Alcohol'); #eje Y
```

### 8.- Diagrama de caja para mostrar la distribucion de la varialbe de Contenido.

```python
fig, ax = plt.subplots()
ax.boxplot(df['Contenido'])
ax.set_title('Distribucion del Cotenido');
```
